```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from urllib.request import urlopen

url="https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%ED%99%8D%EC%9D%B5%EB%8C%80%ED%95%99%EA%B5%90"

page=urlopen(url)
soup=BeautifulSoup(page,"html.parser")
```

```python
soup.select('ul.type01>li>dl>dt>a')[0]
```

```python
news_title=[]
news_paper=[]
```

```python
titles=soup.select('ul.type01>li>dl>dt>a')

for title in titles:
    print(title['title'])
    news_title.append(title['title'])
```

```python
papers=soup.select('ul.type01>li>dl>dd>span')
for paper in papers:
    print(paper.text)
    news_paper.append(paper.text)
```

```python
data1=pd.DataFrame({'paper':news_paper})
```
